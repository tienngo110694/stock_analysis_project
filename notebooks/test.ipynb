```python
import sys
import os
# Ensure we can import the StockDataProvider from the parent directory
sys.path.append(os.path.abspath('..'))

from stock_data_provider import StockDataProvider
import pandas as pd
```

```python
def test_stock_data_provider(stock_code="VNM", start_date="2024-01-01", end_date="2024-03-31", report_date="2022-12-31"):
    """
    A test function to verify that StockDataProvider can fetch all related data.
    """
    print(f"--- Testing StockDataProvider for {stock_code} ---")
    provider = StockDataProvider(stock_code)
    
    # Fetch data
    data = provider.get_comprehensive_data(
        start_date=start_date,
        end_date=end_date,
        report_date=report_date
    )
    
    # Verify results
    for key, df in data.items():
        if not df.empty:
            print(f"✅ {key.replace('_', ' ').title()}: Found {len(df)} rows.")
            # Optionally show first 2 rows of each
            # display(df.head(2))
        else:
            print(f"❌ {key.replace('_', ' ').title()}: No data found.")
            
    return data

# Run the test
all_data = test_stock_data_provider()
```

```python
# Example: Viewing specifically the Financial Ratios
if not all_data['financial_ratios'].empty:
    print("\nExample Financial Ratios:")
    print(all_data['financial_ratios'][['ratioCode', 'value']].head())
else:
    print("No ratios available for display.")
```
