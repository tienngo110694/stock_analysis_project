# Stock Analysis Flow Test

This notebook tests the full integration flow:
1. **AI Processing**: Extracting stock code and data requirements from natural language.
2. **Data Collection**: Fetching real-time and historical data using `vnstock`.
3. **AI Finalization**: Generating a summary based on the collected data.

```python
import os
import json
from dotenv import load_dotenv
from ai_logic import GeminiProcessor
from stock_information_collector import StockInformationCollect

# Load API Keys
load_dotenv()
GEMINI_KEY = os.getenv("GEMINI_API_KEY")

if not GEMINI_KEY:
    print("❌ Error: GEMINI_API_KEY not found in .env")
else:
    print("✅ Environment ready.")
```

## Test: 'Find information of SHB'

```python
ai = GeminiProcessor(GEMINI_KEY)
collector = StockInformationCollect()

user_input = "Find information of SHB"
print(f"User Input: {user_input}")

# Step 1: Analyze Request
summary = await ai.summarize_request(user_input)
print(f"AI Summary: {summary}")

stock_code = summary.get("stock_code")
info_needed = summary.get("info_needed", ["price"])

if stock_code:
    # Step 2: Collect Data
    print(f"Fetching data for {stock_code}...")
    raw_data = await collector.collect_data(stock_code, info_needed)
    
    # Step 3: Final Analysis
    print("Generating final analysis...")
    final_reply = await ai.finalize_analysis(raw_data)
    
    print("\n--- FINAL REPLY ---")
    print(final_reply)
else:
    print("Failed to extract stock code.")
```
